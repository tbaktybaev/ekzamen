```python
from tkinter import *
import math 
from math import*
from tkinter.filedialog import *
import string
import fileinput
pr=Tk()
pr.title('Задачи линейного программирования')
def new_win():
    win = Toplevel(pr)

def close_win():
    win = Toplevel(pr)
    lab = Label(win)
    lab.pack()
    label = tkinter.Label(lab, text="Сохранить файл?")
    label.pack()
    button2 = tkinter.Button(lab, text="Да", command=_save)
    button2.pack()
    button1 = tkinter.Button(lab, text="Нет", command =_close)
    button1.pack()

def _close():
    pr.destroy()

def about():
    win = Toplevel(pr)
    lab = Label(win, text="Это просто программа - пример \n с меню в Tkinter")
    lab.pack()

def _open():
    op = askopenfilename()
    for l in fileinput.input(op):
        txt.insert(END,l)
        
def _save():
    sa = asksaveasfilename()
    letter = txt.get(1.0, END)
    f = open(sa,"w")
    f.write(letter)
    f.close()
    
def open_class():
    class Application(Frame):
        def __init__(self,fr):
            super(Application,self).__init__(fr)
            self.grid()
            self.widget()
            
        def click_revised(self):
    
            from scipy.optimize import linprog

            koef_fun = self.textbox1.get()
            stoim = koef_fun.split(',')
            for i in range(len(stoim)):
                stoim[i] = int(stoim[i])*(-1)
            print(stoim)
            
            koef_fun1 = self.textbox2.get()
            sklad = koef_fun1.split(',')
            for i in range(len(sklad)):
                sklad[i] = int(sklad[i])
            print(sklad)
            
            
            b = self.textbox3.get()
            c = b.split(';')
            matriz = [list(map(int, i.split(','))) for i in c]

            opt = linprog(stoim,matriz ,sklad,
                          method="revised simplex")

            vfunc = ' '.join([f'{-i}*x_{j} +' for j, i in enumerate(stoim)]).rstrip('+') + ' = C'
            print(f'\n\nЦелевая функция: {vfunc}\nОптимальное значение целевой функции: {-opt.fun}\nОптимальные значения переменных решения: {opt.x}')
            
            self.lab = Label(self, text = f'\n\nЦелевая функция: {vfunc}\nОптимальное значение целевой функции: {abs(opt.fun)}\nОптимальные значения переменных решения: {opt.x}' )
            self.lab.grid(row=10,column=0,padx="20",pady="8")
            
            
        def click_simplex(self):
    
            from scipy.optimize import linprog

            koef_fun = self.textbox1.get()
            stoim = koef_fun.split(',')
            for i in range(len(stoim)):
                stoim[i] = int(stoim[i])*(-1)
            print(stoim)
            
            koef_fun1 = self.textbox2.get()
            sklad = koef_fun1.split(',')
            for i in range(len(sklad)):
                sklad[i] = int(sklad[i])
            print(sklad)
            
            
            b = self.textbox3.get()
            c = b.split(';')
            matriz = [list(map(int, i.split(','))) for i in c]

            opt = linprog(stoim,matriz ,sklad,
                          method="simplex")

            vfunc = ' '.join([f'{-i}*x_{j} +' for j, i in enumerate(stoim)]).rstrip('+') + ' = C'
            print(f'\n\nЦелевая функция: {vfunc}\nОптимальное значение целевой функции: {-opt.fun}\nОптимальные значения переменных решения: {opt.x}')
            
            self.lab = Label(self, text = f'\n\nЦелевая функция: {vfunc}\nОптимальное значение целевой функции: {abs(opt.fun)}\nОптимальные значения переменных решения: {opt.x}' )
            self.lab.grid(row=11,column=0,padx="20",pady="8")
        
        
        def click_point(self):
    
            from scipy.optimize import linprog

            koef_fun = self.textbox1.get()
            stoim = koef_fun.split(',')
            for i in range(len(stoim)):
                stoim[i] = int(stoim[i])*(-1)
            print(stoim)
            
            koef_fun1 = self.textbox2.get()
            sklad = koef_fun1.split(',')
            for i in range(len(sklad)):
                sklad[i] = int(sklad[i])
            print(sklad)
            
            
            b = self.textbox3.get()
            c = b.split(';')
            matriz = [list(map(int, i.split(','))) for i in c]

            opt = linprog(stoim,matriz ,sklad)

            vfunc = ' '.join([f'{-i}*x_{j} +' for j, i in enumerate(stoim)]).rstrip('+') + ' = C'
            print(f'\n\nЦелевая функция: {vfunc}\nОптимальное значение целевой функции: {-opt.fun}\nОптимальные значения переменных решения: {opt.x}')
            
            self.lab = Label(self, text = f'\n\nЦелевая функция: {vfunc}\nОптимальное значение целевой функции: {abs(opt.fun)}\nОптимальные значения переменных решения: {opt.x}' )
            self.lab.grid(row=12,column=0,padx="20",pady="8")
            

        def widget(self):
            self.lab=Label(self,text='Введите стоимость каждого n вида кофе(через запятую): ')
            self.lab.grid(row=0,column=0,padx = '20', pady = "8")
            self.lab=Label(self,text='Введите количество ингридиентов на n складе(через запятую): ')
            self.lab.grid(row=2,column=0,padx = '20', pady = "8")
            self.lab=Label(self,text='Введите количество ингридиентов по столбцу n через запятую, cтолбцы разделяйте точкой с запятой:  ')
            self.lab.grid(row=4,column=0,padx = '20', pady = "8")
            
            self.textbox1=Entry(self,width=67,font=("Times New Roman",16))
            self.textbox1.grid(row=1,column=0,padx = '20', pady = "8")
            self.textbox2=Entry(self,width=67,font=("Times New Roman",16))
            self.textbox2.grid(row=3,column=0,padx = '20', pady = "8")
            self.textbox3=Entry(self,width=67,font=("Times New Roman",16))
            self.textbox3.grid(row=5,column=0,padx = '20', pady = "8")
            
            self.btn=Button(self,font=("Times New Roman",12))
            self.btn["text"]="Посчитать оптимальное значение\nцелевой функции двухфазовым симплекс-методом"
            self.btn["command"]=self.click_revised
            self.btn["width"]=45
            self.btn["height"]=3
            self.btn.grid(row=6,column=0)
            
            self.btn=Button(self,font=("Times New Roman",12))
            self.btn["text"]="Посчитать оптимальное значение\nцелевой функции симплекс-методом"
            self.btn["command"]=self.click_simplex
            self.btn["width"]=45
            self.btn["height"]=3
            self.btn.grid(row=7,column=0)
            
            self.btn=Button(self,font=("Times New Roman",12))
            self.btn["text"]="Посчитать оптимальное значение\nцелевой функции методом внутренней точки"
            self.btn["command"]=self.click_point
            self.btn["width"]=45
            self.btn["height"]=3
            self.btn.grid(row=8,column=0)
           




    root=Tk()
    root.title("Производственная задача")
    root.geometry("1000x1000")
    app=Application(root)
    root.mainloop()
    
def open_class_0():
    class Application0(Frame):
        def __init__(self,fr):
            super(Application0,self).__init__(fr)
            self.grid()
            self.widget()
            
        def click(self):
    
            from scipy.optimize import linprog
            import pandas as pd
            file = self.textbox1.get()
            df = pd.read_csv(file,sep=';')
            a,b=df.shape[0],df.shape[1]
            sol=list(df.loc[a-1])
            sol=list(map(lambda x:x*-1,sol))
            ogr=list(df['y'])
            matrix = []
            i=a
            while i!=1:
                j=list(df.loc[a-i])
                matrix.append(j[:-1])
                i-=1
            opt= linprog(sol[:-1], matrix, ogr[:-1], method="revised simplex")
            print(opt)
           
            self.lab = Label(self, text = f'Коэффициенты целевой функции:\n{df.loc[a-1][:-1]}\nОптимальное значение целевой функции: {abs(opt.fun)}\nОптимальные значения переменных решения: {opt.x}' ) 
            self.lab.grid(row=7,column=0,padx="20",pady="8")
            
    
            

        def widget(self):
            self.lab=Label(self,text='Введите путь к файлу: ')
            self.lab.grid(row=0,column=0,padx = '20', pady = "8")
            
            self.textbox1=Entry(self,width=67,font=("Times New Roman",16))
            self.textbox1.grid(row=1,column=0,padx = '20', pady = "8")
            
            self.btn=Button(self,font=("Times New Roman",14))
            self.btn["text"]="Готово"
            self.btn["command"]=self.click
            self.btn["width"]=35
            self.btn["height"]=3
            self.btn.grid(row=6,column=0)
           


    root=Tk()
    root.title("Производственная задача")
    root.geometry("740x550")
    app=Application0(root)
    root.mainloop()

men = Menu(pr)
pr.config(menu=men)
fmen=Menu(men)
men.add_cascade(label='Файл', menu=fmen)
fmen.add_command(label="Новый", command=new_win)
fmen.add_command(label="Открыть", command=_open)
fmen.add_command(label="Сохранить", command=_save)
fmen.add_command(label="Выход", command=close_win)


smen = Menu(men)
men.add_cascade(label='Помощь', menu=smen)
smen.add_command(label="Помощь F1")
smen.add_command(label="О программе", command=about)
nfmen = Menu(fmen)
fmen.add_cascade(label="Импорт", menu=nfmen)
nfmen.add_command(label="картинка")
nfmen.add_command(label="Текст")

vmen = Menu(men)
men.add_cascade(label='Производственная задача', menu=vmen)
vmen.add_command(label='Ввод данных вручную',command=open_class)
vmen.add_command(label='Импорт данных из csv',command=open_class_0)

txt = Text(pr, width=40, height=15, font="12")
txt.pack()
pr.mainloop()
```
